# Merging two data frames

```python
#merging and concatenate

df1 = pd.DataFrame ([['a',1],['b',2]],columns = ['col1','number'])
df2 = pd.DataFrame ([['c',3, 'lion'],['d',4, 'tiger'], ['e',5, 'tiger'],['f',6, 'lion']],columns = ['letter','number', 'animal'])
```

```python
df1
```

```python
df2
```

```python
#lets concatenate row wise
pd.concat([df1,df2], axis =0)
```

```python
pd.concat([df1,df2], axis= 0, ignore_index=True)
```

# Merging part 2

```python
d1 = pd.DataFrame({
                "city":['lucknow', 'Delhi', 'Kanpur', 'Agra', 'Bhubaneswar'],
                "temperature": [42,45,38,41,36]
})
```

```python
 d1
```

```python
d2 = pd.DataFrame({
                "city":['Delhi','Agra','Kanpur'],
                "Humidity":[62,54,60]
})
```

```python
d2
```

```python
#lets merge these two dataframes on the column key city
new_df = pd.merge(d1,d2, on = 'city')
```

```python
new_df
```

```python
new_df2 = pd.merge(d1,d2, on = 'city', how = 'outer')
```

```python
new_df2
```

```python
new_df3 = pd.merge(d1,d2, on = 'city', how = 'left')
```

```python
new_df3
```

```python
#pivot table. (its an alternative of the group by function)
new_df3
```

```python
new_df3.pivot_table(values = 'temperature',index = 'city', aggfunc = 'mean' )
```

```python
new_df3.pivot_table(columns = 'temperature')
```
